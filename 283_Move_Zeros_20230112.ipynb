## 283. Move Zeroes

https://leetcode.com/problems/move-zeroes/

Given an integer array nums, move all 0's to the end of it while maintaining the relative order of the non-zero elements.

Note that you must do this in-place without making a copy of the array.


### Example 1:

Input: nums = [0,1,0,3,12]

Output: [1,3,12,0,0]


### Example 2:

Input: nums = [0]

Output: [0]
 

### Constraints:

* 1 <= nums.length <= 104
* -231 <= nums[i] <= 231 - 1
 

Follow up: Could you minimize the total number of operations done?

```python
# Logic: count ending zero's

def moveZeros(nums):
    n = len(nums)
    print("Starting Array: " + str(nums))

    if (n > 2):
        i = 0
        nonzero = 0
        while (i < n and nums[i] ):
            print("i = " + str(i)) # 0 to 5
            
            if (nums[i] == 0):
                
                # find next non-zero --escaping consecuitive 0's
                for j in range(i,n):
                    print(" "*5 + "j = " + str(j))
                    if (nums[j] != 0):
                        print(" "*5 + "nonzero j = " + str(j))
                        nonzero = j                
                        print(" "*5 + "[x,y] = " + str([nums[i],nums[nonzero]]))
                        nums[i] = nums[nonzero] #swap
                        nums[nonzero] = 0 #swap
                        break # breaking FOR Loop
                i += 1
                print(" "*5 + "Rearranged Array: "  + str(nums))
            
        print("Final Array: " + str(nums))
    elif (n == 2):
        if (nums[1] > 0 and nums[0] == 0):
            nums[0] = nums[1]
            nums[1] = 0 
        print("Final Array: " + str(nums))
    else:
        print("Final Array: " + str(nums))
    
```

```python
arr1 = [0,1,0,3,12]
arr2 = [0]
arr3 = [0,0,0,1,0,5,2,1,2]

moveZeros(arr1)
```
